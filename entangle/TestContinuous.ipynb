```python
import numpy as np
import pyqentangle
```

```python
from scipy.stats import norm
```

```python
fn1 = lambda x1, x2: 1/np.sqrt(2.)*(norm.pdf(x1)*norm.pdf(x2) + norm.pdf((x1-10)/2.)*norm.pdf((x2-15)/3.))
```

```python
modes = pyqentangle.continuous_schmidt_decomposition(fn1, nb_x1=1000, nb_x2=1000, x1_lo=-30., x1_hi=5., x2_lo=-30., x2_hi=5., keep=10)
```

```python
from matplotlib import pyplot as plt
```

```python
x = np.linspace(-30., 5., 100)
```

```python
y1 = modes[1][1](x)
```

```python
y2 = modes[1][2](x)
```

```python
%matplotlib inline
```

```python
plt.plot(x, y1)
```

```python
plt.plot(x, y2)
```
