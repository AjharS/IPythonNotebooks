```python
import numpy as np
from scipy.spatial import Delaunay
from matplotlib import pylab as plt
%matplotlib inline
```

```python
# circular ring
ring = np.array([np.random.normal(loc=1, scale=0.01)*np.array([np.cos(t), np.sin(t)]) for t in np.linspace(0, 2*np.pi, 101)])[1:]
```

```python
plt.plot(ring[:,0], ring[:,1])
```

```python
ring_tri = Delaunay(ring)
```

```python
ring_tri.simplices.shape
```

```python
plt.triplot(ring[:,0], ring[:,1], ring_tri.simplices)
```

```python
from mogu.topology import SimplicialComplex, AlphaComplex
```

```python
ring_c = AlphaComplex(ring, 0.1)
```
