```python
import numpy as np
import keras
from keras.models import Model
from keras.layers import Input, Lambda
```

```python
x = Input((None, 1), dtype='float32')
```

```python
y = Lambda(lambda xx: xx[:, :-1])(x)
```

```python
model = Model(inputs=[x], outputs=[y])
```

```python
model.compile(loss='categorical_crossentropy', optimizer='rmsprop')
```

```python
narr = np.random.uniform(size=(100,10,1))
```

```python
narr
```

```python
model.predict(narr)
```
