```python
import numpy as np
from scipy.spatial import Delaunay
from matplotlib import pylab as plt
%matplotlib inline
```

```python
# circular ring
ring = np.array([np.random.normal(loc=1, scale=0.01)*np.array([np.cos(t), np.sin(t)]) for t in np.linspace(0, 2*np.pi, 101)])[1:]
```

```python
plt.plot(ring[:,0], ring[:,1])
```

```python
ring_tri = Delaunay(ring)
```

```python
ring_tri.simplices.shape
```

```python
ring_tri.simplices
```

```python
plt.triplot(ring[:,0], ring[:,1], ring_tri.simplices)
```

```python
from mogu.topology import AlphaComplex
```

```python
ring_c = AlphaComplex(ring, 0.05)
```

```python
print(ring_c.betti_number(0))
print(ring_c.betti_number(1))
print(ring_c.betti_number(2))
```

```python
sphere = np.array([np.random.normal(loc=1, scale=0.01)*np.array([np.sin(theta)*np.cos(phi), np.sin(theta)*np.sin(phi), np.cos(theta)]) for theta in np.linspace(0.02, np.pi, 51) for phi in np.linspace(0.02, 2*np.pi, 51)])
```

```python
sphere.shape
```

```python
sphere_c = AlphaComplex(sphere, 0.05)
```

```python
sphere_c.simplices
```

```python
len(sphere_c.simplices)
```

```python
print(sphere_c.betti_number(0, eps=0.01))
print(sphere_c.betti_number(1, eps=0.01))
print(sphere_c.betti_number(2, eps=0.01))
```
