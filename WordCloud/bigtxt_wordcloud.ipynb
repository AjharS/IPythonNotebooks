```python
from wordcloud import WordCloud
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
textdict = {
    'summary': 100,
    'text': 50,
    'topic': 40,
    'tf-idf': 40,
    'graph': 20,
    'LDA': 30,
    'LSI': 27,
    'extractive': 85,
    'model': 5,
    'machine learning': 7
}
```

```python
tokens_topick = list(textdict.keys())
total_counts = sum(textdict.values())
tokens_probs = map(lambda token: textdict[token]/float(total_counts), tokens_topick)
generated_tokens = np.random.choice(tokens_topick, p=tokens_probs, size=1000)
```

```python
text = ' '.join(generated_tokens)
```

```python
wordcloud = WordCloud().generate(text)
```

```python
plt.imshow(wordcloud)
```
